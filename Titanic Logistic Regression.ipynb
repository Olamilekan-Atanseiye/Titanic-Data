##  using logistic regression

```python
pd.get_dummies(titanic_cna['Sex'])
#drop one of the colums since when male is 1:female will definately 0
sex=pd.get_dummies(titanic_cna['Sex'],drop_first=True)
sex.head(5)
```

```python
pd.get_dummies(titanic_cna['Embarked'])
#drop one of the colums
embarked=pd.get_dummies(titanic_cna['Embarked'],drop_first=True)
embarked.head(5)
```

```python
pd.get_dummies(titanic_cna['Pclass'])
#drop one of the colums
pclass=pd.get_dummies(titanic_cna['Pclass'],drop_first=True)
pclass.head(5)
```

```python
titanic_cna=pd.concat([titanic_cna,sex,embarked,pclass],axis=1)
```

```python
titanic_cna
```

```python
titanic_cna.drop(['Sex','PassengerId','Name','Ticket','Embarked','Pclass'],axis=1,inplace=True)
titanic_cna
```

## Traning and testing and building a model

```python
x=titanic_cna.drop('Survived',axis=1)
y=titanic_cna['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.3, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
```

```python
logmodel.fit(x_train,y_train)
```

```python
pred=logmodel.predict(x_test)
```

#####   3 methods to check the accuracy of the test
##### ---- clasification_report
##### ---- confusion matrix
##### ---- accuracy_score

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,pred)
```

```python
from sklearn.metrics import confusion_matrix
```
